```python
transactions = pd.read_csv('../data/processed/sales.csv')
transactions.head()
```

```python
Cohort_yr = pd.DataFrame(transactions.groupby('Customer ID')['Year'].min()).reset_index()
Cohort_yr.columns = ['Customer ID', 'Cohort Yr']
Cohort_yr
```

```python
transactions = transactions.merge(Cohort_yr, on='Customer ID').reset_index(drop=True)
transactions = transactions.drop(['Unnamed: 0'],axis=1)
transactions
```

```python
transactions.columns
transactions = transactions[['Customer ID', 'Cohort Yr', 'Customer', 'Type', 'Date', 'Year',
       'Month', 'Document Number', 'Description', 'Qty', 'Unit Price',
       'Sales Total', 'Revenue', 'State']]
transactions
```

```python
def customer_cohort_chart_data(data, x = 'Years', y = 'Sales Total' , cohort_column = 'Cohort Yr'):
    
    sales_yr_cohort = pd.DataFrame(transactions.groupby(['Year','Cohort Yr'])['Sales Total'].sum())
        
    years = sales_yr_cohort['Year'].unique()
    print(years)
    
    sales_by_cohort = {}
    
    for idx, val in enumerate(years):
         sales = sales_yr_cohort[sales_yr_cohort['Cohort Yr']==2015]['Sales Total'].to_list(),
         sales_by_cohort[val] = sales    
            
    return years, sales_by_cohort
```

```python
def plot_customer_cohort_chart(x, y):
    fig, ax = plt.subplots()
    ax.stackplot(x, y.values(), labels=y.keys())
    ax.legend(loc='upper left')
    ax.set_title('Sales by Customer Cohort Year')
    ax.set_xlabel('Year')
    ax.set_ylabel('Sales ($)')
    return plt
```
